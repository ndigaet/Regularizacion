```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def moving_average(a, n=3) :
    ret = np.cumsum(a, dtype=float)
    ret[n:] = ret[n:] - ret[:-n]
    return ret[n - 1:] / n
```

```python
plt.figure(figsize=(15,6)) 

np.random.seed(12) #12
mean = 50
std = 20
data = np.random.normal(mean, std, 100000)
# data = np.array(list(data)*10)
plt.hlines(data.mean(), 0, iterat)

for momentum in [0.9, 0.99, 0.999]:
    
    batch_size = 100
    iterat = int(len(data)/batch_size)
    average_mean = data[:batch_size].mean()
    batch_mean = [average_mean]
    average_means = [average_mean]

    for i in range(iterat-1):
        sample_mean = data[(i+1)*batch_size:(i+2)*batch_size].mean()
        batch_mean.append(sample_mean)
        average_mean = momentum*average_mean + (1-momentum)*sample_mean
        average_means.append(average_mean)

   
    #
    plt.plot(average_means, label=f'Momentum: {momentum}')
    
window = 100
mav = moving_average(batch_mean, n=window)
mav = window*[data.mean()] + list(mav)
plt.plot(mav, label=f'moving Average {window}')
plt.title(f'Normal distribution with mean {mean} and std {std}')
plt.xlabel('Training Iteration')
plt.ylabel('running mean')
plt.legend()
```
